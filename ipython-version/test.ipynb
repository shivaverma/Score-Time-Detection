```python
import cv2
import matplotlib.pyplot as plt
```

```python
def generate_one_frame(frame_number, clip_name):

    cap = cv2.VideoCapture("video/"+clip_name)

    frame_no = frame_number
    cap.set(1,frame_no)
    res, frame = cap.read()
    plt.imshow(frame)
    print(frame.shape)

    cv2.imwrite('data/demo.png', frame)
```

```python
generate_one_frame(23213, "match.mp4")
```

```python
def generate_many_frame(total_frame, clip_name):

    cap = cv2.VideoCapture("video/"+clip_name)
    starting_frame = 10000
    stride = 200
    
    for i in range(total_frame):
    
        frame_no = starting_frame + i*(stride)
        cap.set(1,frame_no)
        res, frame = cap.read()
        cv2.imwrite('clips/frame'+str(i+500)+'.png', frame)   
```

```python
generate_many_frame(500, "match2.mp4")
```
